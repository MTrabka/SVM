```python
import pandas as pd
import numpy as np

df_true = pd.read_csv('True.csv')
df_fake = pd.read_csv('Fake.csv')
```

```python
df_true=df_true.drop(['title','subject','date'], axis = 1)
df_true
```

```python
df_fake=df_fake.drop(['title','subject','date'], axis = 1)
```

```python
df_fake['label']=0
df_true['label']=1
```

```python
df=pd.concat([df_fake, df_true], ignore_index=True)
df
```

```python
X_raw = df['text'].values
y = df['label'].values
```

```python
from sklearn.model_selection import train_test_split #pip install scikit-learn -> w cmd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X_raw)
```

```python
def classification(train_size,kernel):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= train_size, random_state=42)
    model = SVC(kernel = kernel)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return accuracy_score(y_test, y_pred)

train_size = [0.2, 0.4, 0.6, 0.8]
kernels = ['linear', 'poly', 'rbf']

for i in train_size:
    for j in kernels:
        acc=classification(i,j)
        print("Train size = ", i, "kernel =", j, "ACC =", acc)
```
